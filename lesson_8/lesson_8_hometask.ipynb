## *Brest cancer dataset*


```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

from IPython.display import Image
import graphviz
```

```python
cancer = load_breast_cancer()
X, y, labels, features = cancer.data, cancer.target, cancer.target_names, cancer.feature_names
print ('labels:',labels)
print ('features:', features)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

## *DecisionTree*

```python

clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)

print("train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score(X_test, y_test)))

# Візуалізація
graph_viz = sklearn.tree.export_graphviz(clf, out_file=None,feature_names=features, class_names=labels, filled = True)

graph = graphviz.Source(graph_viz)
display(graph)  
```

## *RandomForest*

```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = RandomForestClassifier().fit (X_train, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```

## *Gradient Boosting Decision Trees (GBDT)*


```python
# YOUR_CODE.  create the classifier and train it 
# START_CODE 
clf = GradientBoostingClassifier(learning_rate=0.01,max_depth=3).fit(X_train, y_train)
# END_CODE 

print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```

## *XGBoost*

```python
clf = XGBClassifier()
clf.fit(X_train, y_train)
print("train accuracy= {:.3%}".format(clf.score (X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score (X_test, y_test)))
```
