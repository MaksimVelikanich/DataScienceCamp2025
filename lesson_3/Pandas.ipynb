## Question 1

```python
import pandas as pd 
import numpy as np


def answer_one():
# LOAD ENERGY
    energy = pd.read_excel('/Users/user/Desktop/Camp2025/lesson_3/Energy Indicators.xls', skiprows=17, skipfooter=38)

    energy = energy.drop(energy.columns[[0, 1]], axis=1)

    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

    energy = energy.replace('...', np.NaN)

    energy.iloc[:, 1] *= 1000000

    energy['Country'] = energy['Country'].str.replace(r'\d+', '', regex=True)
    energy['Country'] = energy['Country'].str.replace(r'\(.*?\)', '', regex=True)
    energy['Country'] = energy['Country'].str.replace(r'\s+', ' ', regex=True).str.strip()
    renema_countries = {
        "Republic of Korea": "South Korea",
        "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"
    }
    energy['Country'] = energy['Country'].replace(renema_countries)



    # LOAD GDP
    GDP = pd.read_csv('/Users/user/Desktop/Camp2025/lesson_3/world_bank.csv', skiprows=4)
    GDP = GDP.iloc[:, :-1]

    rename_columns = {
        "Korea, Rep.": "South Korea",
        "Iran, Islamic Rep.": "Iran",
        "Hong Kong SAR, China": "Hong Kong"
    }

    GDP['Country Name'] = GDP['Country Name'].replace(rename_columns)

    years = [str(year) for year in range(2006, 2016)]
    GDP = GDP[['Country Name'] + years]

    # LOAD SCIMEN
    ScimEn = pd.read_excel('/Users/user/Desktop/Camp2025/lesson_3/scimagojr country rank 1996-2023.xlsx')
    ScimEn_top15 = ScimEn[ScimEn['Rank'] <= 15]


    merged = pd.merge(ScimEn_top15, energy, on='Country', how='inner')
    merged = pd.merge(merged, GDP, left_on='Country', right_on='Country Name', how='inner')
    merged.drop(columns=['Country Name'], inplace=True)

    # marging data
    columns_to_keep = [
            'Country', 'Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
            'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita',
            '% Renewable'
        ] + years
    res = merged[columns_to_keep].set_index('Country')


    return res

answer_one()


```

## Question 2

```python
def answer_two():
    Top15 = answer_one()
    years = [str(year) for year in range(2006, 2016)]
    avgGDP = Top15[years].mean(axis=1).sort_values(ascending=False)
    return avgGDP
answer_two()
```

## Question 3

```python
def answer_three():
    Top15 = answer_one()
    avgGDP = answer_two()
    country = avgGDP.index[5]
    return Top15.loc[country]['2015'] - Top15.loc[country]['2006']

answer_three()
```

## Question 4

```python
def answer_four():
    Top15 = answer_one()
    Top15['Ratio'] = Top15['Self-citations'] / Top15['Citations']
    country = Top15['Ratio'].idxmax()
    ratio = Top15['Ratio'].max()
    return country, ratio
answer_four()
```

## Question 5

```python
def answer_five():
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    res = Top15.sort_values('Population', ascending=False)
    return res.index[2]

answer_five()
```

## Question 6

```python
def answer_six():
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable docs per Capita'] = Top15['Citable documents'] / Top15['Population']
    return Top15['Citable docs per Capita'].corr(Top15['Energy Supply per Capita'])

answer_six()
```

## Question 7

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict  = {
                  'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    

    if 'Country' in Top15.columns:
        Top15.set_index('Country', inplace=True)
    Top15['Continent'] = Top15.index.map(ContinentDict)
    Top15.dropna(subset=['Continent'], inplace=True)
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    res = Top15.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])
    return res

answer_seven()
```
