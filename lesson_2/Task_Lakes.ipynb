# **Task about lakes**

* Generate 20 random variable from 1 to 100
* Draw the plot of the sequence
* Lets assume each point represent the height and so all plotting is 2d mountains.
* Then consider the unlimited rain from above - cavities become lakes full of water.
* Determine the deepest lake

_Based on my own code from the test task_

```python
import matplotlib.pyplot as plt
import numpy as np

def generateRandomData(): # Functiom "generateRandomData" generates random data of 20 integers between 0 and 100
    np.random.seed(2023)
    y = np.random.randint(0, 101, 20)
    return y

def deepestLake(heights): # Function "deepestLake" finds the deepest lake in the given data
    max_depth = 0
    lake_start, lake_end = 0, 0 # initialize the start and end of the lake
    
    for i in range(1, len(heights) - 1): 
        left_max = max(heights[:i]) # find the maximum height to the left of the current height
        right_max = max(heights[i+1:]) # find the maximum height to the right of the current height
        depth = min(left_max, right_max) - heights[i] # calculate the depth of the lake
        
        if depth > max_depth: # if the depth of the lake is greater than the maximum depth found so far
            max_depth = depth 
            lake_start = np.argmax(heights[:i+1]) # find the index of the start of the lake
            lake_end = np.argmax(heights[i:]) + i # find the index of the end of the lake
     
    return max_depth, lake_start, lake_end

def visualize(heights): # Function "visualize" visualizes the given data
    x = np.arange(len(heights))
    plt.plot(x, heights)

    max_depth, lake_start, lake_end = deepestLake(heights)
    
    plt.plot(x[lake_start:lake_end+1], heights[lake_start:lake_end+1], color='red')

    plt.show()

heights = generateRandomData()
visualize(heights)
result = deepestLake(heights)
print("The deepest Lake:", result[0])
```
