## *Descriptive Statistics*


```python
print(df.describe())

for col in df.select_dtypes(include='object').columns:
    print('-'* 30)
    print(col)
    print('')
    print(f"\n{col}:\n", df[col].value_counts(normalize=True))

print("\nOutcome counts:")
print(df['outcome'].value_counts())

print("\nOutcome by percent:")
print(df['outcome'].value_counts(normalize=True) * 100)

print("\nMean values of numerical variables by outcome:")
print(df.groupby('outcome').mean(numeric_only=True))

```

## *Data Visualization*


```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_columns = ['age','gender', 'vehicle_type', 'race', 'education', 'driving_experience']

for col in categorical_columns:
    plt.figure(figsize=(2, 2))
    filtered = df[df[col] != 0] 
    
    
    sns.countplot(data=filtered, x=col, hue=col, legend=False, 
                   palette='Set1')
    
    
    plt.title(col)
    plt.show()

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='education', hue='outcome')
plt.xticks(rotation=45)
plt.show()



```

```python
corr = df.iloc[:,[7,8,10,11,13,15,16,17,18]].corr()
sns.heatmap(corr, annot=True, linewidths=1, cmap='rocket_r')
```

## *Conclusion*


*After analyzing the car insurance dataset , several important patterns were discovered:*
* Only about 30% of customers submit insurance claims, indicating a highly imbalanced target variable.
* People aged 16 to 25 are more likely to submit claims than older age groups.
* Clients with secondary or incomplete higher education are more likely to file claims than those with higher education.
* Drivers with less than 5 years of experience are significantly more likely to be involved in insurance incidents.
* Women submit claims slightly less often than men.
* Individuals with lower credit scores have a higher probability of claiming insurance.
* Drivers with higher annual mileage are less likely to file claims.
* Clients with a history of violations are more likely to be claimants.

**These insights demonstrate that behavioral, social, and financial features strongly influence claim probability. This information can help insurance companie optimize pricing, risk assessment, and personalized policy offerings.**
